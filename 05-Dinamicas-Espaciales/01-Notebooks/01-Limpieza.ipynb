## **DELITOS**

```python
delitos = pd.read_csv(r"C:\Heri\GitHub\.Semester\Geospatial-Data-Analysis\07-Trabajo\Delitos\Municipal-Delitos-2015-2025_ene2025.csv", encoding='latin1')
```

```python
delitos = delitos[delitos['Entidad'] == 'Ciudad de México']
```

```python
# drop Entidad column
delitos = delitos.drop(columns=['Entidad', 'Clave_Ent'])
```

```python
# drop where Municipio == 'No Especificado'
delitos = delitos[delitos['Municipio'] != 'No Especificado']
```

```python
# keep before 2025
delitos = delitos[delitos['Año'] < 2025]
```

```python
# save memory
delitos = delitos.drop(columns=['Municipio'])
```

```python
# change column names
delitos = delitos.rename(columns={'Cve. Municipio': 'Clave'})
```

```python
delitos
```

```python
delitos.to_csv(r'delitos-cdmx.csv', index=False)
```

# **MUN.SHP**

```python
    
mun = gpd.read_file(r'C:\Heri\GitHub\.Semester\Geospatial-Data-Analysis\07-Trabajo\09_ciudaddemexico\conjunto_de_datos\09mun.shp')
```

```python
mun = mun.drop(columns=['CVE_ENT', 'CVE_MUN'])
```

```python
# drop the firt str from CVEGEO
mun['CVEGEO'] = mun['CVEGEO'].str[1:]
```

```python
mun
```

```python
# change the name of the column
mun = mun.rename(columns={'CVEGEO': 'Clave', 'NOMGEO': 'Municipio'})
```

```python
mun.crs
```

```python
# convert to EPSG:6372
mun = mun.to_crs(epsg=6372)
```

```python
mun.crs
```

```python
# save to gpkg
mun.to_file(r'mun-cdmx.gpkg', layer='mun', driver='GPKG')
```

## **POBLACIÓN**

```python
censo = pd.read_csv(r'C:\Heri\GitHub\.Semester\Geospatial-Data-Analysis\07-Trabajo\Censo de Poblacion y Vivienda\iter_09_cpv2020\conjunto_de_datos\conjunto_de_datos_iter_09CSV20.csv', encoding='utf-8-sig')
```

```python
censo
```

```python
# we only want the population of the municipalities
poblacion=censo[['MUN', 'NOM_MUN','LOC','POBTOT', 'POBFEM', 'POBMAS']]
```

```python
poblacion
```

```python
censo['NOM_MUN'].unique()
```

```python
# it says the total population of the mun
poblacion=poblacion[poblacion['LOC'] == 0]
```

```python
poblacion
```

```python
# drop the rowns where MUN == Total de la entidad blabla
poblacion = poblacion[poblacion['MUN'] != 0]
```

```python
poblacion.drop(columns=['LOC'], inplace=True)
```

```python
poblacion
```

```python
# MUN from 2 to 002
poblacion['MUN'] = poblacion['MUN'].astype(str).str.zfill(3)
```

```python
# add a 9 before the MUN
poblacion['MUN'] = '9' + poblacion['MUN']
```

```python
# change the names of the columns
poblacion.rename(columns={'MUN': 'Clave', 'NOM_MUN': 'Municipio', 'POBTOT': 'Poblacion Total', 'POBFEM': 'Poblacion Femenina', 'POBMAS': 'Poblacion Masculina'}, inplace=True)
```

```python
poblacion
```

```python
poblacion.to_csv(r'poblacion-cdmx.csv', index=False)
```
