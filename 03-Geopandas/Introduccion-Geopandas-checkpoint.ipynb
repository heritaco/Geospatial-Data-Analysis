# INTRODUCCIÓN A GEOPANDAS

GeoPandas es una librería de Python para manejar datos geoespaciales en el modelo vectorial. Su clase principal son los GeoDataFrame que heredan su estructuran de los pandas.DataFrame, pero permite relacionar cada fila de datos a un objeto geográfico. Además hay otros metadatos asociados, principalmente el sistema de coordenadas proyectadas.

Documentación: http://geopandas.org

Depende de varias librerías, por lo tanto, su instalación puede llegar a ser problemática si se instala incorrectamente las dependencias. La recomendación es usar la fuente conda-forge con el comando:


Las librerías mas importantes de las que depende es:
* Pandas: Manejo y análisis de datos
* Fiona: Leer y escribir diferentes formatos SIG (.shp, .gpkg, .geoJSON, etc)
* GDAL: Leer y escribir diferentes formatos SIG (es una dependencia de Fiona)
* pyproj: Manejo de proyeciones cartográficas y transformación de coordenadas
* numpy: Computo cientifico

Una librería opcional que mejora la eficiencia de las operaciones en geopandas es:
* rtrees: Utiliza una estructura de datos con el mismo nombre para representar relaciones y operaciones de objetos geográficos. GeoPandas lo usa como respaldo si PyGeos no está instalado.
PyGEOS: Es más rápido y se utiliza si está disponible en tu instalación.

# Administrar proyecciones

Para acceder a la información de la proyeción se usa el atributo ".crs"

```python
municipios.crs
```

Para cambiar la proyeción se debe usar el método ".to_crs()" especificando el código EPSG del sistema de referencias de coordendas al que se desean transformar los datos, y asignarlo a una variable (puede ser una nueva o la misma).

Opciones:
**Opcion 1** (pasar el diccionario)
> municipios.to_crs({'proj':'longlat', 'datum'='WGS84', 'no_defs'=True})

**Opcion 2** (especificar el EPSG)
> municipios.to_crs(epsg=4326)

**Opción 3** (pasar el crs de otra capa)
> municipios.to_crs(otro_GeoDataFrame.crs)

```python
municipios = municipios.to_crs(4326)
municipios.crs
```

Cuando el archivo de entrada no tiene un crs específicado se puede definir usando el método "set_crs()". Si el archivo ya tiene un crs y se usa el metodo "set_crs" se generará un error.

### Barra de escala

Se puede agregar barras de escala al generar los mapas con la función *ScaleBar* de la librería  from **matplotlib_scalebar.scalebar**. 

```python
fig, ax = plt.subplots(figsize=(10, 15), dpi=100)
municipios.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.1)
scalebar = ScaleBar(1,  # Factor de escala (1 si las unidades ya están en metros)
                    units="m", 
                    dimension="si-length",  # Sistema internacional
                    location="lower right",  # Ubicación
                    scale_loc="bottom")  # Ubicación del texto de escala
ax.add_artist(scalebar)
ax.set_title("Municipios de México")
ax.set_axis_off()
plt.show()
```

Asegurate que las unidades estén en metros, en caso contrario debes transformar a tu crs a coordenadas proyectadas.

```python
municipios = municipios.to_crs(6372)
fig, ax = plt.subplots(figsize=(10, 15), dpi=100)
municipios.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.1)
scalebar = ScaleBar(1,  # Factor de escala (1 si las unidades ya están en metros)
                    units="m", 
                    location="lower left",
                    scale_loc="bottom")
ax.add_artist(scalebar)
ax.set_title("Municipios de México")
ax.set_axis_off()
plt.show()
```

### Mapas base

También es posible agregar imágenes georreferenciadas de fondo, para ello es necesario (instalar e) importar la librería **contextily**, y usar la funcion *ctx.add_basemap*, estableciendo la fuente adecuada:

```python
import contextily as ctx
```

```python
fig, ax = plt.subplots(figsize=(10, 15), dpi=200)
municipios.boundary.plot(ax=ax, color="red", edgecolor="red", linewidth=0.5)
# Agregar un mapa base con contextily (por ejemplo, OpenStreetMap)
ctx.add_basemap(ax, 
                source=ctx.providers.OpenStreetMap.Mapnik, 
                crs=municipios.crs.to_string())  # Asegura de que el CRS sea el mismo
ax.set_title("Municipios de México")
ax.set_axis_off()
plt.show()
```

# Construir un GeoDataFrame

Construir un nuevo GeoDataFrame sigue la misma estructura que un panda.DataFrame, es decir, introducir un conjunto de datos (filas) con un conjunto de atributos (columnas). Pero, además debe introducirse una columna definiendo el objeto geográfico de cada dato, lo cual se hace con clases de Shapely, así como definir el sistema de referencias de coordenadas 

```python
from shapely.geometry import Point, LineString, MultiPoint, MultiLineString, Polygon, MultiPolygon
```

```python
df = pd.DataFrame(np.random.random(size=[3,3]),columns=['A','B','C'])
df
```

```python
p1 = Point([0,0])
p2 = Point([0,1])
p3 = Point([1,1])
p1
```

```python
geodf = gpd.GeoDataFrame(df, geometry=[p1, p2, p3], crs=4326)
geodf.plot()
```
